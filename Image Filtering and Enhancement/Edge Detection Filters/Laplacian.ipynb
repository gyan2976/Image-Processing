```python
# loading image
img0 = cv2.imread('/content/drive/MyDrive/Computer Vision/Image Processing/Data/kid.jpg')

# converting to gray scale
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)

# convolute with proper kernels
laplacian = cv2.Laplacian(img,cv2.CV_64F)

plt.figure(figsize =(15,15)),plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.figure(figsize =(15,15)),plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])


plt.show()
```

```python
# loading image
img0 = cv2.imread('/content/drive/MyDrive/Computer Vision/Image Processing/Data/Shraddha.jpg')

# converting to gray scale
gray = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)

# convolute with proper kernels
laplacian = cv2.Laplacian(img,cv2.CV_64F)

plt.figure(figsize =(15,15)),plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.figure(figsize =(15,15)),plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])


plt.show()
```

```python
from google.colab.patches import cv2_imshow

#load image
image = cv2.imread("/content/drive/MyDrive/Computer Vision/Image Processing/Data/kid.jpg")
#convert to gray image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#detect gradients, edges
lap_edges = cv2.Laplacian(gray_image, cv2.CV_64F)
#convert all -ve pixels to positives
lap_edges = np.uint8(np.absolute(lap_edges))
cv2_imshow(lap_edges)
cv2.waitKey(0)
```
