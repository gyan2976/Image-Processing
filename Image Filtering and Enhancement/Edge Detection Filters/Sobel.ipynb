#**Vertical (1,0)**

```python
# loading image
vertical_1 = cv2.imread('/content/drive/MyDrive/Computer Vision/Image Processing/Data/kid.jpg')

# converting to gray scale
gray = cv2.cvtColor(vertical_1, cv2.COLOR_BGR2GRAY)

# remove noise
img = cv2.GaussianBlur(gray,(3,3),0)

# convolute with proper kernels
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)  # x

plt.figure(figsize=(8,10)),plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])

plt.figure(figsize=(8,10)),plt.subplot(2,2,2),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])


plt.show()
```

#**Vertical + Horizontal**

```python
from google.colab.patches import cv2_imshow

ddept=cv2.CV_16S
scale=1

img_1 = cv2.imread('/content/drive/MyDrive/Computer Vision/Image Processing/Data/kid.jpg')
gray = cv2.cvtColor(img_1, cv2.COLOR_BGR2GRAY)
x = cv2.Sobel(gray, ddept, 1,0, ksize=3, scale=1)   # Horizontal Edge (1,0)
y = cv2.Sobel(gray, ddept, 0,1, ksize=3, scale=1)   # Vertical Edge (0,1)
absx= cv2.convertScaleAbs(x)
absy = cv2.convertScaleAbs(y)
edge = cv2.addWeighted(absx, 0.5, absy, 0.5,0)
cv2_imshow(edge)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
