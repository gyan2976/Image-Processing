#**Using OpenCV**

```python
import imutils
from google.colab.patches import cv2_imshow

# read an image as input using OpenCV
image = cv2.imread(r"/content/drive/MyDrive/Computer Vision/Image Processing/Data/Shraddha.jpg")

Rotated_image = imutils.rotate(image, angle=45)
Rotated1_image = imutils.rotate(image, angle=90)

# angle 45
cv2_imshow(Rotated_image)

# angle 90
cv2_imshow(Rotated1_image)

cv2.waitKey(0)
```
