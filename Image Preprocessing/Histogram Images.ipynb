```python
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/drive/MyDrive/Computer Vision/Image Processing/Data/Shraddha.jpg', -1)
cv2_imshow(img)

color = ('b','g','r')
for channel,col in enumerate(color):
    histr = cv2.calcHist([img],[channel],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Histogram for color scale picture')
plt.show()
```

```python
from google.colab.patches import cv2_imshow

gray_img = cv2.imread('/content/drive/MyDrive/Computer Vision/Image Processing/Data/Shraddha.jpg', cv2.IMREAD_GRAYSCALE)
cv2_imshow(gray_img)
hist = cv2.calcHist([gray_img],[0],None,[256],[0,256])
plt.hist(gray_img.ravel(),256,[0,256])
plt.title('Histogram for gray scale picture')
plt.show()
```
