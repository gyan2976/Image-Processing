**Use mask image (Draw circle, rectangle, line, etc)**

```python
img1 = Image.open('/content/drive/MyDrive/Computer Vision/Image Processing/Data/Sea.jpg')      # Big picture
img2 = Image.open('/content/drive/MyDrive/Computer Vision/Image Processing/Data/Shraddha.jpg') # Small picture
```

```python
mask_im = Image.new("L", im2.size, 0)
draw = ImageDraw.Draw(mask_im)
draw.ellipse((140, 50, 260, 170), fill=255)
hh = img1.copy()
hh.paste(img2, (0, 0), mask_im)
plt.imshow(hh)
plt.show()
```

```python
mask_im_blur = mask_im.filter(ImageFilter.GaussianBlur(10))
back_img = img1.copy()
back_img.paste(img2, (0, 0), mask_im_blur)
plt.imshow(back_img)
plt.show()
```
